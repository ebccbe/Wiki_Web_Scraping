# Covid - 19 Forecast and analysis

Covid - 19 has been spreading rapidly in the state of Tamil Nadu, India. 

We will analyze the timeline and will forecast new Covid - 19 positive cases in Tamil Nadu.

```python
from opencage.geocoder import OpenCageGeocode

import pandas as pd
data = pd.read_html("https://en.wikipedia.org/wiki/COVID-19_pandemic_in_Tamil_Nadu")
print(len(data))
```

```python
new_header= ['Date','Date2','NA', '# of cases', '# of deaths']
```

```python
print(new_header)
```

```python
tn.columns = new_header
```

```python
tn
```

```python
tn.drop(["NA"], axis = 1, inplace = True)
tn
```

```python
tn.drop(tn.tail(1).index,inplace=True)
tn.drop(tn.head(1).index,inplace=True)
tn
```

```python
tn.drop(["Date2"], axis=1, inplace = True)
tn
```

```python
tn[['No of cases','Change']] = tn["# of cases"].str.split("(",expand=True)
tn
```

```python
tn.drop(["# of cases"], axis = 1, inplace = True)
tn
```

```python
tn['Change'] = tn['Change'].map(lambda x: x.lstrip('+').rstrip('%)'))
tn
```

```python
tn[['No of deaths','DeathChange']] = tn["# of deaths"].str.split("(",expand=True)
tn
```

```python
tn.drop(["# of deaths"], axis = 1, inplace = True)
tn
```

```python
tn.drop(["DeathChange"], axis = 1, inplace = True)
tn
```

```python
tn.describe()
```

```python
tn["Change"].describe()
```

```python
tn.groupby("Change").count()
```

```python
tn['Change'].mask(tn['Change'] == 'n.a.', '0', inplace=True)
tn['Change'].mask(tn['Change'] == '=', '0', inplace=True)
tn.groupby("Change").count()
```

```python
tn["No of deaths"] = tn["No of deaths"].fillna(0)
tn
```

```python
list_lat = []   # create empty lists

list_long = []



for index, row in tn_dist.iterrows(): # iterate over rows in dataframe



    City = row['District']
    
    query = str(City)+', Tamil Nadu'

    results = geocoder.geocode(query)   
    lat = results[0]['geometry']['lat']
    long = results[0]['geometry']['lng']

    list_lat.append(lat)
    list_long.append(long)

# create new columns from lists    

tn_dist['lat'] = list_lat   

tn_dist['lon'] = list_long
```
